# **Line plot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=[2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,6,7,8]
plt.plot(x,y)
plt.show()
```

```python
import pandas as pd
var=pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=var)
plt.show()
```

```python
df=sns.load_dataset('penguins').head(20)
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=df)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',size=50)
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',style='sex')
plt.show()
```

```python
sns.lineplot(x='bill_length_mm',y='bill_depth_mm',data=df,hue='sex',style='sex',palette='Accent',markers=['o','>'])
plt.show()
```

# **Heatmap**
accept only 2d data

```python
import numpy as np
```

```python
var=np.linspace(1,10,20).reshape(4,5)
var                                #range k andar data mil sake
```

```python
sns.heatmap(var)
```

```python
sns.heatmap(var,vmin=0,vmax=12,cmap='PuOr',annot=True)
```

```python
var=np.linspace(1,10,10).reshape(2,5)
font={"fontfamily":"monospace",'fontweight':"bold",'color':'k'}
ar=np.array([['a0','a1','a2','a3','a4'],['b0','b1','b2','b3','b4']])
x=sns.heatmap(var,vmin=0,vmax=10,cmap='PuOr',annot=ar,fmt="s",annot_kws=font,lw=5,linecolor='r',cbar=False,xticklabels=False)
x.set(xlabel='abc',ylabel='pqr')
sns.set(font_scale=2)
plt.show()
```
