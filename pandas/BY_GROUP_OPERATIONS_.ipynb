```python
def grou(df,idx,col):
  if 80<=df[col].loc[idx]<=90:
    return '80-90'
  elif 50<=df[col].loc[idx]<80:
    return '50-80'
  else:
    return 'other'
g= df.groupby(lambda x:grou(df,x,'temperature'))
```
