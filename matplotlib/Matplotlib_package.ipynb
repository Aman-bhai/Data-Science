```python
import matplotlib.pyplot as plt
```

***plot function***

```python
x=[2,3,4,5,6,8]
y=[3,5,68,9,0,8]
y1=[1,12,31,41,51,6]
plt.plot(x,y1,color='green',label='first111')
plt.plot(x,y,marker='*',mfc='blue',ms='20',color='red',lw=2,ls='dotted',label='first')
plt.title("graph b\w x and y")
plt.xlabel("x-axis")
plt.ylabel('y-label')
plt.grid()
plt.legend()
```

***scatter plot***

```python
x=[2,3,4,5,6,8]
y=[3,5,68,9,0,8]
y1=[1,12,31,41,51,6]
plt.scatter(x,y1,color='green',label='first111')
plt.scatter(x,y,marker='*',color='red',label='first')
plt.title("graph b\w x and y")
plt.xlabel("x-axis")
plt.ylabel('y-label')
plt.grid()
plt.legend()
```

```python
import pandas as pd
```

```python
df=pd.read_csv('Iris_data_sample.xlsx - Iris_data.csv',na_values=['###','??'])
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
x=df.SepalLengthCm
y=df.SepalWidthCm
z=df.Species
Z=df.PetalLengthCm
```

```python
plt.plot(x,color='red',marker='+',mfc='yellow',mec='yellow')
plt.plot(y,color='blue',marker='+',mfc='black',mec='magenta')
```

```python
plt.xlabel('SepalLengthCm')
plt.grid
plt.ylabel('SepalWidthCm')
plt.title('line plot')
plt.plot(x,y)
```

```python
df.head()
```

```python
plt.xlabel('Species')
plt.ylabel('Sepallength')
plt.title('histogram represent the sepallength and Species')
plt.bar(z,x)
```

```python
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.title('histogram represent the sepallength and SepalWidthCm')
plt.bar(x,y)
```

```python
plt.xlabel('species')
plt.ylabel('SepalLengthCm')
plt.title('histogram represent the SepalWidthCm and sepallength')
plt.barh(y,x)
```

```python
plt.xlabel('species')
plt.ylabel('SepalLengthCm')
plt.title('histogram represent the sepallength of various species')
plt.barh(z,x)
```

```python
plt.title('pie plot of sepallength')

plt.pie(df.SepalLengthCm.head(),labels=z.head(),autopct='%2f');
```

```python
dic={'family':'serif','size':20,'weight':'bold','color':'blue'}
dic1={'family':'serif','size':25,'weight':'bold','color':'Magenta'}
dic2={'family':'serif','size':30,'weight':'bold','color':'Red'}
```

```python
plt.xlabel('SepalLengthCm',fontdict=dic)
plt.grid
plt.ylabel('SepalWidthCm',fontdict=dic1)
plt.title('scatter plot',fontdict=dic2)
plt.scatter(x,y)
```

```python
plt.axes(projection='3d').plot3D(x.head(), y.head(), Z.head(), 'red')
plt.xlabel('sepallength')
plt.ylabel('sepalwidth')
plt.title('3d plot')
plt.grid()
plt.show()
```
