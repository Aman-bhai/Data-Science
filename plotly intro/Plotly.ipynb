```python
import numpy as np
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```


1.   plotly creates interactive visualizations in the form of HTML files
2.   can't able to handle live data source
2.   Dash is used to create data based dashboard



```python
delieveries=pd.read_csv("deliveries.csv")
matches=pd.read_csv("matches.csv")
```

```python
delieveries
```

```python
matches
```

```python
df=delieveries.merge(matches,left_on="match_id",right_on="id")
df
```

```python
bestFifty=df.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(50).index.tolist()
new_df=df[df["batsman"].isin(bestFifty)]
new_df
```

```python
#calculation of strike rate
# sr=no.of run scored/no.of balls played

runs=new_df.groupby("batsman")["batsman_runs"].sum()
balls=new_df.groupby("batsman")["batsman_runs"].count()

sr=(runs/balls)*100
sr=sr.reset_index()
sr
```

```python
 # calculating average
 #avg=(total no. of runs/no.of outs)

out=df[df["player_dismissed"].isin(bestFifty)]
nouts=out["player_dismissed"].value_counts()
avg=runs/nouts
avg=avg.reset_index()
avg.rename(columns={'index':"batsman",0:"avg"},inplace=True)

avg=avg.merge(sr,on="batsman")
avg
```

## Scatter plot

```python
trace=go.Scatter(x=avg["avg"],y=avg["batsman_runs"],mode="markers",text=avg["batsman"],marker={"color":"green","size":16},name="v kohli")

data=[trace]

layout=go.Layout(title="Batsman Avg vs SR",xaxis={"title":"Batsman Average"},
                 yaxis={"title":"Batsman Strike rate(sr)"})

fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename="scatter.html")
```

## Line Chart

```python
single=df[df["batsman"]=="V Kohli"]
performance=single.groupby("season")["batsman_runs"].sum().reset_index()
performance
```

```python
def batsman_comp(*name):
  data=[]
  for i in name:
    single=df[df["batsman"]==i]
    performance=single.groupby('season')["batsman_runs"].sum().reset_index()

    trace=go.Scatter(x=performance["season"],y=performance["batsman_runs"],mode="lines+markers",name=i)

    data.append(trace)
    layout=go.Layout(title="Batsman Record Comparator",xaxis={"title":"Season"},
                     yaxis={"title":"Runs"})

    fig=go.Figure(data=data,layout=layout)
    pyo.plot(fig,filename="Multilinesusin.html")

batsman_comp("V kohli","RG Sharma","DA Warner","MS Dhoni")
```

## Bar Graph





```python
bestTen=df.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10).index.tolist()

bestTen_df=df[df["batsman"].isin(bestTen)]
```

```python
trace=go.Bar(x=bestTen_df["batsman"],y=bestTen_df["batsman_runs"])

data=[trace]

layout=go.Layout(title="Bar graph",
                 xaxis={"title":"Batsman"},
                 yaxis={"title":"Batman_runs"})
fig=go.Figure(data=data,layout=layout)

pyo.plot(fig,filename="Bar.html")
```

### There are 2 types of Bar Graphs


1.   Nested Bar Graph
2.   Stacked Bar Graph
3.   Overlayed Bar Graph



```python
iw=bestTen_df.groupby(["batsman","inning"])["batsman_runs"].sum().reset_index()
mask=iw["inning"]==1
mask2=iw["inning"]==2
one=iw[mask]
two=iw[mask2]

one.rename(columns={"batsman_runs":"1st Innings"},inplace=True)
two.rename(columns={"batsman_runs":"2nd Innings"},inplace=True)

final=one.merge(two,on="batsman")[["batsman","1st Innings","2nd Innings"]]
final
```

```python
trace1=go.Bar(x=final["batsman"],y=final["1st Innings"],name="1st Innings",marker={"color":"red"})
trace2=go.Bar(x=final["batsman"],y=final["2nd Innings"],name="2nd Innings",marker={"color":"blue"})

data=[trace1,trace2]
layout=go.Layout(title="Inning wise Scores",
                  xaxis={"title":"Batsman"},
                  yaxis={"title":"Runs"})
                  # barmode="overlay")
                  # barmode="stack")
fig=go.Figure(data=data,layout=layout)

pyo.plot(fig,filename="Bars.html")
```

## Histogram

```python
x=delieveries.groupby("batsman")["batsman_runs"].count()>150
x=x[x].index.tolist()

new=delieveries[delieveries["batsman"].isin(x)]

runs=new.groupby("batsman")["batsman_runs"].sum()
balls=new.groupby("batsman")["batsman_runs"].count()

sr=(runs/balls)*100
sr=sr.reset_index()
sr
```

```python
trace=go.Histogram(x=sr["batsman_runs"],name="Strike rate variation",xbins={"size":10})
data=[trace]

layout=go.Layout(title="Strike rate analysis",
                 xaxis={"title":"Strike rates"})

fig=go.Figure(data=data,layout=layout)
pyo.plot(fig,filename="histo.html")
```

## HEATMAPS

```python
six=delieveries[delieveries["batsman_runs"]==6]
six=six.groupby(["batting_team","over"])["batsman_runs"].count().reset_index()
six
```

```python
trace=go.Heatmap(x=six["batting_team"],y=six["over"],z=six["batsman_runs"])

data=[trace]

layout=go.Layout(title="Six Heatmap")

fig=go.Figure(data=data,layout=layout)

pyo.plot(fig,filename="Heatmap.html")
```

```python
dots=delieveries[delieveries["batsman_runs"]==0]
dot=dots.groupby(["batting_team","over"])["batsman_runs"].count().reset_index()
from plotly import tools

traces=go.Heatmap(x=six["batting_team"],y=six["over"],z=six["batsman_runs"].values.tolist())
traces2=go.Heatmap(x=dots["batting_team"],y=dots["over"],z=dots["batsman_runs"].values.tolist())

fig=tools.make_subplots(rows=1,cols=2,subplot_titles=["6's","0's"],shared_yaxes=True)
fig.append_trace(traces,1,1)
fig.append_trace(traces2,1,2)

pyo.plot(fig,filename="multi_heatmap.html")
```
