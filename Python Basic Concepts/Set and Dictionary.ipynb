# Set

A set is a collection which is both unordered and unindexed.

Sets are written with curly brackets.

```python
a={'aman','kumar',4,8.53}
print(a)
```

```python
print(type(a))
```

```python
l=['aman','kumar',4,8.53]
print(l)
s=set(l)
print(s)
print(type(s))
```

```python
# Duplicate values will be ignored:
s = {"a","b","c","d","a","b"}

print(s)
```

***add function***

```python
# Add an item to a set, using the add() method:
s= {"a","b","c"}

s.add("hello world")

print(s)
```

***update function***

```python
s= {"a","b","c"}
s1={"hello world","spartaa"}
s.update(s1)
print(s)
```

***remove functions***

```python
s.remove("spartaa")
s
```

***discard function***

```python
s.discard('hello world')
s
```

***pop function***

```python
s.pop()
s
```

***clear function***

```python
s.clear()
s
```

# **Dictionary**

```python
d={1:"rahul",2:'ram',3:'shyam',4:'tomar',5:'soni',7:'aman',8:'amit',9:'sharma',10:'singh'}
print(type(d))
print(d[2])
```

```python
d1={'aman':7,'amiar':8,'amar':20,'male':['aman','amiair']}
```

***key and value function***

```python
print(d1.keys())
print(d1.values())
```

```python
d1['aman']='student'
d1
```

```python
d2={(5,9):{2,3}}
print(d2)
```

***clear function***

```python
print(d2.clear())
print(d2)
```

***delete***

```python
del d2
print(d2)
```

***min and max functions***

```python
d2={3:5,6:8,5:8,7:0,2:7}
print ("Dictionary -->", d2)
print("minimum value in d2 is",min(d2.values()))
print("maximum  value in d2 is", max(d2.values()))
```

***the key which have min value and max value***

```python

print('minimum key',min(d2,key=d2.get))
print('maximum key',max(d2,key=d2.get))
```
