##Tuple

1.   Unchangeable
2.   Ordered
3.   Tuple items are ordered, unchangeable, and allow duplicate values.

Tuple items are indexed, the first item has index [0], the second item has index [1] etc.





```python
t1 = ("apple", "banana", "cherry")
t2 = (1, 5, 7, 9, 3)
t3 = (True, False, False)
print(t1)
print(t2)
print(t3)
```

```python
t1=(1,2,3,4)
t2=(10,11,12,13)
t=t1+t2
print(t)
```

```python
f = ("a", "b", "c")
f = f * 2

print(f)
```

```python
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)

print(x)
print(y)
```

```python
T=[1,2,3,'hello','world']
T=list(T)
T[2]=80
del T[0]
print(T)

print(type(T))
T=tuple(T)
print(T)
print(type(T))
```

```python
T=("aman","raj","19/01/2001",1,2,3,True,"hello",("rad","rahul",2,6,9,False,5.5))
T1=(1,2,3,4)
print(T)
print(len(T))
print(T[0])
print(type(T))
print(T+T1)
print(1 in T)
print("Length of Tuple:",len(T))
print("Maximum element in Tuple:",max(T1))
print("Minimum element in Tuple:",min(T1))
print("Sorting in Tuple:",sorted(T1))
print("No. of  appearnace of particular element in Tuple:",T1.count(2))
print("Index of element in Tuple:",T1.index(3))
```
