# List Items

List items are ordered, changeable, and allow duplicate values.

List items are indexed, the first item has index [0], the second item has index [1] etc.

```python
li=[1,2,3.3,'aman',True]
```

```python
print(li[-1])
print(li[0])
print(li[1])
print(li[2])
print(li[3])
print(li[4])
print(type(li))
```

```python
list1 = ["a", "b", "c"]
list2 = [1, 2, 3]

list3 = list1 + list2

print(list3)
```

```python
li=[1,2,3.3,'aman',True]
L2=[1,2,3.3,'bhai',False]
li+L2
print(li+L2)
```

```python
l = list(("apple", "banana", "cherry")) # note the double round-brackets
print(l)
```

```python
l = ["apple", "banana", "cherry", "melon", "mango"]
print(l[-4:-1])

if "apple" in l:
  print("Hello")
```

***sort function***

```python
l2=[2,3,5,446,76,87,4,5657,8]
l2.sort()
```

```python
l2=[2,3,5,446,76,87,4,5657,8]
l2.sort(reverse=True)
```

```python
l = ["orange", "mango", "kiwi", "pineapple", "banana"]
l.sort(reverse = True)
print(l)
```

```python
# Perform a case-insensitive sort of the list:
l = ["banana", "Orange", "Kiwi", "cherry"]
l.sort(key = str.lower)

print(l)
```

```python
# Sort the list based on how close the number is to 50:

def functions2(n):
  return abs(n - 50)

l = [100, 50, 65, 82, 23]
l.sort(key =functions2)
print(l)
```
