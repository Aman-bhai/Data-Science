***Relation Operator***

```python
a=3
b=2
print(a>3)    # greater than operator
print(a<3)    #less than operator
print(a==3)
print(b<=5)     # less than or equal to
print(b>=5)     # greater than or equal to
```

***Logical Operator***

```python
a,b =3,5
print( a==3 and b==5)   # and operator  return true if both expression is true
print( a==5 and b==5)
print(a==3 or b==3 )         # or operator return true if any expression is true
print(a==3 or b==5 )
print(a==6 or b==6 )
print(a!=3 )              # not operator
```
