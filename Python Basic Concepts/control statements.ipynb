```python
num=int(input())
for x in range(1,11):             # range() gives the range of interger
  print(num,"*",x,"=",num*x)
```

```python
c=0
num=int(input())
for x in range(1,num):
  if num%x==0:
    c+=1
if c>=2:
  print("number is not prime")
else:
  print('number is prime')
```

```python
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

***Jumping statements***

```python
# Continue to the next iteration if i is 3:
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
for x in range(6):
  if x == 3: break      # break statement
  print(x)
else:
  print("Finally finished!")
```
